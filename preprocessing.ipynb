# Data preprocessing

### Panoramica

In questo notebook, finalizziamo la preparazione del dataset. Le operazioni principali saranno:
1. caricamento dataset pulito
2. separazione tra feature (X) e variabile target (y)
3. splitting: training (70%), validation (15%) e test (15%)
4. scaling delle feature

### Caricamento dataset

```python
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_csv("data/cleaned.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head(10)
```

Rimuoviamo la colonna 'id' che non serve per il training del modello

```python
df = df.drop('id', axis=1)
```

### Separazione feature e target

```python
y = df['vote_average']
X = df.drop('vote_average', axis=1)
```

```python
print(f"shape features: {X.shape}")
print(f"shape target: {y.shape}")
```

### Suddivisione in Training, Validation e Test Set


Applichiamo uno split in due passaggi per ottenere i tre set di dati desiderati (70% train, 15% validation, 15% test).

Usiamo random_state=42 per garantire la riproducibilit.

```python
TEST_SIZE = 0.15
VALIDATION_SIZE = 0.15
RANDOM_STATE = 42
```

```python
from sklearn.model_selection import train_test_split
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE
)
```

```python
validation_split_ratio = VALIDATION_SIZE / (1 - TEST_SIZE)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp,
    test_size=validation_split_ratio,
    random_state=RANDOM_STATE
)
```

Check sulle dimensioni:

```python
print(f"Training set:   X_train -> {X_train.shape}, y_train -> {y_train.shape}")
print(f"Validation set: X_val   -> {X_val.shape}, y_val   -> {y_val.shape}")
print(f"Test set:       X_test  -> {X_test.shape}, y_test  -> {y_test.shape}")
```

### Analisi della varianza
Molte di queste, in particolare quelle derivate dal TF-IDF, potrebbero avere una varianza molto bassa (essere quasi costanti), aggiungendo rumore senza fornire molto segnale predittivo. Per creare un modello pi robusto e semplice, applico una tecnica di feature selection univariata. L'obiettivo  rimuovere automaticamente le feature che hanno una varianza al di sotto di una certa soglia.

per fare questa analisi,  meglio scalare le feature interamente in [0,1] (scaling tempoeraneo)

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import MinMaxScaler

scaler_for_variance = MinMaxScaler()
X_train_minmax = scaler_for_variance.fit_transform(X_train)
threshold = 0.007

selector = VarianceThreshold(threshold=threshold)
selector.fit(X_train_minmax)

features_to_keep_mask = selector.get_support()
features_to_keep = X_train.columns[features_to_keep_mask]
features_to_drop = X_train.columns[~features_to_keep_mask]

print(f"--- Selezione Feature basata sulla Varianza (soglia={threshold}) ---")
print(f"Numero di feature originali: {X_train.shape[1]}")
print(f"Numero di feature mantenute: {len(features_to_keep)}")
print(f"Numero di feature scartate: {len(features_to_drop)}")
```

```python
print("Feaures to keep: "+str(list(features_to_keep)))
```

```python
print("Feaures to drop: "+str(list(features_to_drop)))
```

```python
X_train = X_train[features_to_keep]
X_val = X_val[features_to_keep]
X_test = X_test[features_to_keep]
```

Check sulle dimensioni:

```python
print(f"Training set:   X_train -> {X_train.shape}, y_train -> {y_train.shape}")
print(f"Validation set: X_val   -> {X_val.shape}, y_val   -> {y_val.shape}")
print(f"Test set:       X_test  -> {X_test.shape}, y_test  -> {y_test.shape}")
```

z### Scaling sulle feature numeriche (StandardScaler)

Le feature numeriche rimanenti si trovano ancora su scale molto diverse (es. "release_year" vs. "log_popularity"). Per i modelli sensibili alla scala come le regressioni lineari,  fondamentale standardizzare i dati. Utilizzo "StandardScaler" per trasformare le feature in modo che abbiano media 0 e deviazione standard 1.

```python
columns_to_scale = [
    'release_year',
    'log_popularity',
    'log_vote_count'
] + [col for col in X_train.columns if col.startswith('tfidf_')]
columns_to_scale
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = X_train.copy()
X_val_scaled = X_val.copy()
X_test_scaled = X_test.copy()

X_train_scaled[columns_to_scale] = scaler.fit_transform(X_train[columns_to_scale])

X_val_scaled[columns_to_scale] = scaler.transform(X_val[columns_to_scale])
X_test_scaled[columns_to_scale] = scaler.transform(X_test[columns_to_scale])
```

```python
print(X_train_scaled.head())
```

Non mi sono limitato a creare feature, ma ho anche implementato una fase di feature selection per ridurre il rumore. Ho usato VarianceThreshold per rimuovere 31 feature quasi costanti, principalmente derivate dal TF-IDF, rendendo il modello finale pi semplice e potenzialmente pi generalizzabile.

Salviamo tutti i set di dati (X e y per train, validation, test)

```python
print(f"Training set:   X_train -> {X_train.shape}, y_train -> {y_train.shape}")
print(f"Validation set: X_val   -> {X_val.shape}, y_val   -> {y_val.shape}")
print(f"Test set:       X_test  -> {X_test.shape}, y_test  -> {y_test.shape}")
```

```python
import os

output_dir = 'processed_data'
base_path = os.getcwd()
full_output_path = f"{base_path}/{output_dir}"

X_train_scaled.to_csv(f"{full_output_path}/X_train.csv", index=False, encoding='utf-8')
X_val_scaled.to_csv(f"{full_output_path}/X_val.csv", index=False, encoding='utf-8')
X_test_scaled.to_csv(f"{full_output_path}/X_test.csv", index=False, encoding='utf-8')

y_train.to_csv(f"{full_output_path}/y_train.csv", index=False, encoding='utf-8')
y_val.to_csv(f"{full_output_path}/y_val.csv", index=False, encoding='utf-8')
y_test.to_csv(f"{full_output_path}/y_test.csv", index=False, encoding='utf-8')
```
